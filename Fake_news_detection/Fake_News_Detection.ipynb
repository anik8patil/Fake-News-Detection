## Fake news detection

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import re
import string
```

```python
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
```

```python
df_fake.head(10)
```

```python
df_true.head(10)
```

```python
df_fake["class"] = 0
df_true["class"] = 1
```

```python
df_fake.shape, df_true.shape
```

```python
df_fake_manual_testing = df_fake.tail(10)
for i in range(23501,23491, -1):
    df_fake.drop([i], axis=0, inplace=True)
df_true_manual_testing = df_true.tail(10)
for i in range(21416,21406, -1):
    df_true.drop([i], axis=0, inplace=True)
```

```python
df_manual_testing = pd.concat([df_fake_manual_testing, df_true_manual_testing], axis=0)
df_manual_testing.to_csv("manual_testing.csv")
```

```python
df_marge = pd.concat([df_fake, df_true], axis=0)
df_marge.head(10)
```

```python
df = df_marge.drop(["title", "subject", "date"], axis=1)
df.head(10)
```

```python
df = df.sample(frac=1)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ", text)
    text = re.sub('https?://\S+!www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```

```python
df.head(10)
```

```python
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_tarin, y_test = train_test_split(x,y, test_size = .25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectrorization = TfidfVectorizer()
xv_train = vectrorization.fit_transform(x_train)
xv_test = vectrorization.transform(x_test)
```

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_tarin)
```

```python
LR.score(xv_test, y_test)
```

```python
pred_LR = LR.predict(xv_test)
```

```python
print(classification_report(y_test, pred_LR))
```

#### Manual Testing

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(word_drop) 
    new_x_test = new_def_test["text"]
    new_xv_test = vectrorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)

    return print((output_lable(pred_LR[0])))
                                                                                                              
```
